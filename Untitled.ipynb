```python
import pandas as pd
import pyteomics.mzml
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#PSM over time

pdx6449_rep1 = pd.read_csv("PXD006449/pdx6449_rep1", sep='\t')
pdx6449_rep2 = pd.read_csv("PXD006449/pdx6449_rep2", sep='\t')
```

```python
#split the time column so you get one that's just minutes
pdx6449_rep1['retention_time'] = pdx6449_rep1['retention_time'].astype(str)
pdx6449_rep1["temp_minute"] = pdx6449_rep1["retention_time"].str.split("\.")
pdx6449_rep1.loc[:, 'minute'] = pdx6449_rep1['temp_minute'].map(lambda x: x[0])
pdx6449_rep1 = pdx6449_rep1.drop(columns="temp_minute")
pdx6449_rep1[["minute"]] = pdx6449_rep1[["minute"]].apply(pd.to_numeric)
pdx6449_rep1= pdx6449_rep1.sort_values("minute")


#split the time column so you get one that's just minutes
pdx6449_rep2['retention_time'] = pdx6449_rep2['retention_time'].astype(str)
pdx6449_rep2["temp_minute"] = pdx6449_rep2["retention_time"].str.split("\.")
pdx6449_rep2.loc[:, 'minute'] = pdx6449_rep2['temp_minute'].map(lambda x: x[0])
pdx6449_rep2 = pdx6449_rep2.drop(columns="temp_minute")
pdx6449_rep2[["minute"]] = pdx6449_rep2[["minute"]].apply(pd.to_numeric)
pdx6449_rep2= pdx6449_rep2.sort_values("minute")
```

```python
linedf_rep1 = pdx6449_rep1['minute']
linedf_rep1 = linedf_rep1.value_counts().rename_axis('minute').reset_index(name='counts_rep1')
linedf_rep1 = linedf_rep1.sort_values("minute")
linedf_rep1 = linedf_rep1.rename({'minute':'minute_rep1'}, axis=1)

linedf_rep2 = pdx6449_rep2['minute']
linedf_rep2 = linedf_rep2.value_counts().rename_axis('minute').reset_index(name='counts_rep2')
linedf_rep2 = linedf_rep2.sort_values("minute")
linedf_rep2 = linedf_rep2.rename({'minute':'minute_rep2'},axis=1)
```

```python

a4_dims = (30, 8.27) #dimensions for bigger plot
fig, ax = plt.subplots(figsize=a4_dims)
sns.lineplot(data=linedf_rep1, x="minute_rep1", y="counts_rep1")
sns.lineplot(data=linedf_rep2, x="minute_rep2", y="counts_rep2")
ax.legend(['rep1','rep2','rep3'],fontsize='20')
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
plt.title('PDX006449',fontsize='20')
plt.ylabel('Number of PSMs', fontsize=30)
plt.xlabel('Minute', fontsize=30)
# plt.savefig('Figures/supplemental_figure2A.png', dpi=600)
plt.ylim(0,100)
plt.show()
```
