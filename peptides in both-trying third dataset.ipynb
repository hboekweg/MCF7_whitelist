```python
formated_df = df2[['Peptide','protein','best_tR','m_z', 'Incubator_1','Incubator_2','Incubator_3','Incubator_4']]
formated_df

import statistics
def average_intesnities(row):
    return statistics.mean([row['Incubator_1'],row['Incubator_2'],row['Incubator_3'],row['Incubator_4']])
    
formated_df['intensity'] = formated_df.apply(average_intesnities, axis=1)


#sort by intesnity
formated_df = formated_df.sort_values('intensity', ascending = False)
# formated_df.head()
```
